```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

ancestry_colors = {
    'European': '#0173B2',
    'East Asian': '#DE8F05',
    'Native American': '#029E73',
    'African': '#CC78BC',
    'South Asian': '#CA9161',
    'Ancient Hunter-Gatherer': '#949494',
    'Oceanian': '#ECE133',
    'West Asian': '#56B4E9'
}

# Create legend handles
legend_handles = [mpatches.Patch(color=color, label=name) 
                  for name, color in ancestry_colors.items()]

# Add legend to your plot
plt.legend(handles=legend_handles, loc='best')

# Or create a standalone legend figure
fig, ax = plt.subplots(figsize=(6, 4))
ax.axis('off')
ax.legend(handles=legend_handles, loc='center', frameon=False, fontsize=12)
plt.tight_layout()
plt.show()
```
